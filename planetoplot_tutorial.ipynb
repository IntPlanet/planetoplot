## Use in a python script

Line-by-line description
* Script header
* Import `pp` object from `ppclass`
* Define a `pp` object named `ex`
* Define attributes
  - netCDF file to read
  - variable to read
  - value on y axis
  - value on z axis
* Get data + Make plot

```python
#! /usr/bin/env python
from ppclass import pp
ex = pp()
ex.file = 'diagfired.nc'
ex.var = 'temp'
ex.quiet = True
ex.y = 15.
ex.z = 75.
ex.getplot()
```

* Customize your plot

```python
ex.title = r'Hovmoller plot for $T_{75km}$'
ex.xcoeff = 24.
ex.xlabel = 'Time (hours)'
ex.ylabel = 'Altitude (km)'
ex.vmin = 130.
ex.vmax = 170.
ex.div = 20
ex.fmt = '%.0f'
ex.colorbar = "spectral"
ex.plot() # remake plot
```
